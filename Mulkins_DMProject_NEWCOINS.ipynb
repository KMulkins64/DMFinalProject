```python
import pandas as pd
coins = pd.read_csv('coins.csv')
```

```python
len(coins)
```

```python
import numpy as np
newCoins = pd.DataFrame()
for i in range(1, len(coins)+1):
  newArray = []
  attr = coins.values[i-1:i]
  newArray = pd.DataFrame(attr, columns=coins.columns)
  newArray = pd.DataFrame(pd.concat([newArray]*11, ignore_index = True))
  value_mat = newArray.iloc[:, 4:15]
  newVals = np.diag(value_mat)
  newArray['Value'] = newVals
  col_names = newArray.columns[4:15]
  newArray['Mint_State'] = col_names
  newCoins = newCoins.append(newArray[['Denom', 'Year', 'Mintage', 'Mint_State', 'Value']], ignore_index=True)
```

```python
newCoins = newCoins[newCoins['Value'] != -999]
```
