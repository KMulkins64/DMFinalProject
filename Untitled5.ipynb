```python
# @title
import pandas as pd
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error

df = pd.DataFrame(coinsAGood[['Denom', 'Year', 'Value']])

# Encode categorical variables to numeric using one-hot encoding
df_encoded = pd.get_dummies(df, columns=['Denom'], drop_first=True)

# Split the data into training and testing sets
X_train, X_test, y_train, y_test = train_test_split(
    df_encoded.drop('Value', axis=1),
    df_encoded['Value'],
    test_size=0.2,
    random_state=42
)

# Train a Random Forest model
model = RandomForestRegressor(random_state=42)
model.fit(X_train, y_train)

# Predict on the test set
predictions = model.predict(X_test)

# Calculate the residuals (prediction errors)
residuals = y_test - predictions

# Define a threshold for outliers (you can adjust this value)
threshold = 3

# Identify outliers based on prediction error
outliers = (abs(residuals) > threshold)

# Print the rows containing outliers
print("Rows with outliers:")
print(df.loc[X_test[outliers].index])
```

```python
# @title
indicies = df.loc[X_test[outliers].index].index
indicies = np.array(indicies)
indicies
```

```python
# @title
for i in indicies:
  coinsAGood = coinsAGood.drop(i)
```

```python
import pandas as pd

df = pd.DataFrame(coinsAGood)

# Calculate IQR for the 'Value' column
Q1 = df['Value'].quantile(0.25)
Q3 = df['Value'].quantile(0.75)
IQR = Q3 - Q1

# Define a lower and upper bound for outliers
lower_bound = Q1 - 1.5 * IQR
upper_bound = Q3 + 1.5 * IQR

# Identify outliers based on IQR
outliers = (df['Value'] < lower_bound) | (df['Value'] > upper_bound)

# Print the rows containing outliers
print("Rows with outliers based on IQR:")
print(df[outliers])
```

```python
indicies = df[outliers].index
indicies = np.array(indicies)
indicies
```

```python
from sklearn.preprocessing import LabelEncoder
lencMTState = LabelEncoder()
coins["Mint_State"] = lencMTState.fit_transform(coins["Mint_State"])
coins["Mint_State"]
```

```python
from sklearn.preprocessing import LabelEncoder
lencDenom = LabelEncoder()
coins["Denom"] = lencDenom.fit_transform(coins["Denom"])
coins["Denom"]
```

```python
coins['log_Value'] = np.log(coins['Value'])
```

```python
coins
```

```python
# With what value you can estimate the generalization error?
from sklearn.model_selection import train_test_split
X = coins.iloc[:,0:4]
Y = coins.log_Value

x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size= 0.25, random_state=0)
```

```python
X
```

```python
# Use the Linear Regression Model
from sklearn.linear_model import LinearRegression

# Define a linear regression model
reg = LinearRegression()

# Train and fit the training data
reg.fit(x_train , y_train)

# We will measure how well our linear regression works in our x_test split
# predictions are the estimations
log_predictions = reg.predict(x_test)
predictions = np.exp(log_predictions)
```

```python
from sklearn.metrics import mean_squared_error

# To compare models
mseModel1 = mean_squared_error(y_test, predictions)
```

```python
mseModel1
```

```python
coefficients = reg.coef_
intercept = reg.intercept_

print(f'Coefficients: {coefficients}')
print(f'Intercept: {intercept}')
```

```python
import matplotlib.pyplot as plt

plt.scatter(y_test, predictions)
plt.xlabel('Actual Values')
plt.ylabel('Predicted Values')
plt.title('Actual vs. Predicted Values in Multiple Linear Regression')
plt.show()
```
